## Now, we're going to determine the planet density

Write a function to solve for the planet density in $\frac{g}{cm^3}$. Use the values that you found earlier for $M_p$ (Planet Mass) and $R_p$ (Planet Radius) to calculate it. Remember: You'll need to convert the units!

$$\rho= \frac{3M_p}{4 \pi R_p^3}$$


**Some Helpful Info:**

$R_\oplus = 6.371 \times 10^8 \hspace{2mm} cm$<br>
$M_\oplus= 5.972 \times 10^{27} \hspace{2mm} grams$


If you need some help getting started with functions, here is a great page that shows what a function looks like, how to define and call it:

https://www.w3schools.com/python/python_functions.asp

```python
#Create your density function here 
def Density(m, r): 
    M=m*5.972e27
    R=r*6.371e8
    rho=(3*M)/(4*np.pi*R**3)
    return rho

rho=Density(Mp,Rp)
print(rho), 'g/cm^3')
```

```python
letter='a'
for i in range(8): 
    if i==0:
        print('     Name            Mass             Radius            Density')
        print(' ')
    
    else: 
        letter=chr(ord(letter)+1)
        Name='Trappist 1-'+letter
        print(Name, '    ', Mp[i-1],'M_earth    ',Rp[i-1], 'R_earth    ', np.round(rho [i-1],3), 'g/cm^3    ')
        print(' ')
```

## Let's talk about planet differentiation 

First, we'll need a few terms. You'll also see ExoPlex return these values as well, so make sure to take note!

Core Mass Fraction (CMF):

- The percent of the planet that is core (aka iron) by mass

Core Radius Fraction (CRF):
- The percent of the planet that is core (aka iron) by mass

### With that out of the way we can talk about this a little more

When planets form they're made up of a bunch of different elements that are all mixed up. The cool thing about materials is that over long time scales (like on the scale of millions of years!) most materials exhibit fluid-like behaviour. This means that just like how oil and vinegar in a salad dressing separate out planets do the same thing. The heavier elements like iron (Fe) sink to the bottom and lighter elements "float" on top like silicon (Si) and Magnesium (Mg) the things that rocks are made out of. See the figure below:

 

### Where does the CMF come in??

Well, how much heavy stuff (i.e. iron) you have versus the amount of light stuff (rock: Si, Mg) in a planet can tell you a lot about the planet. For example: Earth has a CMF=.3 (30%), which means that it has a lot more mantle or rock compared to iron. However, if you compare Earth's CMF to Mercury's you'll find that Mercury has a CMF=.7 (70%). It's mosting made up of iron! There are also other weird planets like 55 Cnc e that make astronomers scratch their heads because it has a CMF of 0.01 (1% !!!) (https://exoplanets.nasa.gov/exoplanet-catalog/7005/55-cancri-e/). We can talk more about 55 Cnc e if you're interested, but all you need to know is that most planets have CMF~30%. So, it's super weird. Below is just a slice of how you can have different CMFs and what they look like:
