## Now, we're going to determine the planet density

Write a function to solve for the planet density in $\frac{g}{cm^3}$. Use the values that you found earlier for $M_p$ (Planet Mass) and $R_p$ (Planet Radius) to calculate it. Remember: You'll need to convert the units!

$$\rho= \frac{3M_p}{4 \pi R_p^3}$$


**Some Helpful Info:**

$R_\oplus = 6.371 \times 10^8 \hspace{2mm} cm$<br>
$M_\oplus= 5.972 \times 10^{27} \hspace{2mm} grams$


If you need some help getting started with functions, here is a great page that shows what a function looks like, how to define and call it:

https://www.w3schools.com/python/python_functions.asp

```python
#Create your density function here 
def Density(m, r): 
    M=m*5.972e27
    R=r*6.371e8
    rho=(3*M)/(4*np.pi*R**3)
    return rho

print(Density(Mp,Rp), 'g/cm^3')
```
